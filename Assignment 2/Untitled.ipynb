```python
def cleandata(df):

    # Calculate the percentage of null values in each column
    null_percentages = df.isnull().mean()

    # Drop columns with null percentage greater than 50%
    df = df.loc[:, null_percentages < 0.5]
    
    # Convert the binary label indicating bankruptcy (True/False) to 1 and 0
    df['bankrupt'] = df['bankrupt'].astype(int)
    
    df["feat_4"] = df["feat_4"].fillna(df["feat_4"].median())
    df["feat_5"] = df["feat_5"].fillna(df["feat_5"].median())
    df["feat_8"] = df["feat_8"].fillna(df["feat_8"].median())
    df["feat_9"] = df["feat_9"].fillna(df["feat_9"].median())
    df["feat_12"] = df["feat_12"].fillna(df["feat_12"].median())
    df["feat_13"] = df["feat_13"].fillna(df["feat_13"].median())
    df["feat_15"] = df["feat_15"].fillna(df["feat_15"].median())
    df["feat_16"] = df["feat_16"].fillna(df["feat_16"].median())
     
    df["feat_17"] = df["feat_17"].fillna(df["feat_17"].median())
    df["feat_19"] = df["feat_19"].fillna(df["feat_19"].median())
    df["feat_20"] = df["feat_20"].fillna(df["feat_20"].median())
    df["feat_21"] = df["feat_21"].fillna(df["feat_21"].median())
    df["feat_23"] = df["feat_23"].fillna(df["feat_23"].median())
    df["feat_24"] = df["feat_24"].fillna(df["feat_24"].median())
    df["feat_26"] = df["feat_26"].fillna(df["feat_26"].median())
    df["feat_27"] = df["feat_27"].fillna(df["feat_27"].median())
 
    df["feat_28"] = df["feat_28"].fillna(df["feat_28"].median())
    df["feat_29"] = df["feat_29"].fillna(df["feat_29"].median())
    df["feat_31"] = df["feat_31"].fillna(df["feat_31"].median())
    df["feat_32"] = df["feat_32"].fillna(df["feat_32"].median())
    df["feat_33"] = df["feat_33"].fillna(df["feat_33"].median())
    df["feat_34"] = df["feat_34"].fillna(df["feat_34"].median())
    df["feat_35"] = df["feat_35"].fillna(df["feat_35"].median())
    df["feat_38"] = df["feat_38"].fillna(df["feat_38"].median())
   
    df["feat_40"] = df["feat_40"].fillna(df["feat_40"].median())
    df["feat_41"] = df["feat_41"].fillna(df["feat_41"].median())
    df["feat_42"] = df["feat_42"].fillna(df["feat_42"].median())
    df["feat_43"] = df["feat_43"].fillna(df["feat_43"].median())
    df["feat_44"] = df["feat_44"].fillna(df["feat_44"].median())
    df["feat_45"] = df["feat_45"].fillna(df["feat_45"].median())
    df["feat_46"] = df["feat_46"].fillna(df["feat_46"].median())
    df["feat_47"] = df["feat_47"].fillna(df["feat_47"].median())
     
    df["feat_48"] = df["feat_48"].fillna(df["feat_48"].median())
    df["feat_50"] = df["feat_50"].fillna(df["feat_50"].median())
    df["feat_51"] = df["feat_51"].fillna(df["feat_51"].median())
    df["feat_53"] = df["feat_53"].fillna(df["feat_53"].median())
    df["feat_54"] = df["feat_54"].fillna(df["feat_54"].median())
    df["feat_55"] = df["feat_55"].fillna(df["feat_55"].median())
    df["feat_57"] = df["feat_57"].fillna(df["feat_57"].median())
    df["feat_59"] = df["feat_59"].fillna(df["feat_59"].median())
 
    df["feat_61"] = df["feat_61"].fillna(df["feat_61"].median())
    df["feat_62"] = df["feat_62"].fillna(df["feat_62"].median())
    df["feat_63"] = df["feat_63"].fillna(df["feat_63"].median())
    df["feat_64"] = df["feat_64"].fillna(df["feat_64"].median())
    
    df["bankrupt"] = df["bankrupt"].fillna(df["bankrupt"].median())
    
    return df
```

```python
bankrupt_df = cleandata(df)
print("Missing values: \n", bankrupt_df.isnull().sum())
```

```python
# Step 2: Split the data
from sklearn.model_selection import train_test_split

X=bankrupt_df.drop('bankrupt', axis=1)#features
y=bankrupt_df['bankrupt']#target

# Split the data into training and testing sets with 80% of the data for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,train_size=0.8, random_state=42)
```

```python
# Print the shapes of the training and testing sets
print("Training set shape: ", X_train.shape)
print("Testing set shape: ", X_test.shape)
```

```python
from imblearn.under_sampling import RandomUnderSampler

# Resample target vector using RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer object
imputer = SimpleImputer(strategy='mean')  # or 'median', 'most_frequent', 'constant'

# Fit the imputer on the data and transform it
X_resampled_filled = imputer.fit_transform(X_resampled)
X_test = imputer.fit_transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.01]
}

# Train GradientBoostingClassifier model using GridSearchCV
gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_resampled_filled,y_resampled)
grid_search = GridSearchCV(gb, param_grid, cv=5)
grid_search.fit(X_resampled_filled, y_resampled)

# Select best hyperparameters
best_gb = grid_search.best_estimator_
print("Best hyperparameters: ", grid_search.best_params_)
```

```python
# Print the cross-validation score
print('Cross-validation score:', grid_search.best_score_)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Evaluate model on testing set
y_pred = best_gb.predict(X_test)

# Compute metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy: {:.2f}".format(accuracy))
print("Precision: {:.2f}".format(precision))
print("Recall: {:.2f}".format(recall))
print("F1-score: {:.2f}".format(f1))
```

```python
# Compute feature importance
importance = best_gb.feature_importances_
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': importance})
feature_importance = feature_importance.sort_values('Importance', ascending=False)

# Select top 10 features
top_features = feature_importance.iloc[:10]
print("Top 10 most important features: \n", top_features)

```
